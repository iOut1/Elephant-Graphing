```
# Visualization and data libraries #
import pandas as pd
import matplotlib.pyplot as plt

# Sklearn tool libraries#
from sklearn.preprocessing import PolynomialFeatures

# AI model libraries #
from sklearn.linear_model import LinearRegression
```

```
# Read data file #
Data = pd.read_csv('/work/Elephant Data.csv')
```

```
# Delete rows with no data as most animals with no tusks have been poached #
Data = Data.dropna()
```

```
# Graph data relationship of the age of an elephant and the length of its tusks #
plt.scatter(Data['Estimated age (years)'], Data['Tusk length (cm)'])
```

```
# Create tool for a more accurate prediction #
poly = PolynomialFeatures(degree=3, include_bias=False)
```

```
# Restructure age data #
poly = poly.fit_transform(Data[['Estimated age (years)']])
```

```
# Create model #
model = LinearRegression()
```

```
# Train model #
model.fit(poly, Data[['Tusk length (cm)']])
```

```
# Create predictions on itself #
predictions = model.predict(poly)
```

```
# Graph actual data and predicted data #
plt.scatter(Data['Estimated age (years)'],Data['Tusk length (cm)'])
plt.scatter(Data['Estimated age (years)'], predictions, color='red')
```
